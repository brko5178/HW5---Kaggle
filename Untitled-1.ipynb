```python
def load_image(iid, image_dir=train_dir):
    path = os.path.join(train_dir, iid + ".tif")
    image = cv2.imread(path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    return image
```

```python
TRAIN_SIZE_PER_LABEL=20000
negative = train_labels[train_labels['label'] == 0].sample(TRAIN_SIZE_PER_LABEL)
positive = train_labels[train_labels['label'] == 1].sample(TRAIN_SIZE_PER_LABEL)
neg_and_pos = pd.concat([negative, positive], axis=0).reset_index(drop=True)
train_df_small = neg_and_pos.sample(frac=1).reset_index(drop=True)
```

```python
X = np.array([load_image(i) for i in train_df_small['id']])
y = train_df_small['label'].values
```

This dataset is huge and is difficult to work with, so a smaller subset of the data will be useful. If results seems good with this smaller subset it may justify giving the CNN a larger dataset to train on.

Model and reasoning

```python
m1 = keras.Sequential([
    layers.Input(shape=(96, 96, 3)),
    layers.Rescaling(1./255),
    layers.Cropping2D(cropping=32),
    layers.Conv2D(32, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(128, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(1, activation='sigmoid')
])
m1.compile(optimizer=Adam(learning_rate=0.0001), loss='binary_crossentropy', metrics=['accuracy', 'auc'])
m1.summary()
```

```python
es1 = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
h1 = m1.fit(X, y, validation_split=0.2, epochs=50, callbacks=[es1])
```

```python
test_files = os.listdir(test_dir)
test_ids = [f.split('.')[0] for f in test_files]
test_X = np.array([load_image(i, image_dir=test_dir) for i in test_ids])
```

```python
results = m1.predict(test_X)
results = results.reshape(-1)
results.shape
```

```python
results_df = pd.DataFrame({
    'id': test_ids,
    'label': results.tolist(),
})
```

```python
def graph_train_hist(h):
    # Graph of training/validation loss and accuracy.
    loss = h.history['loss']
    val_loss = h.history['val_loss']
    accuracy = h.history['accuracy']
    val_accuracy = h.history['val_accuracy']
    auc = h.history['auc']
    val_auc = h.history['val_auc']
    epochs = range(1, len(loss) + 1)
    
    plt.figure(figsize=(18, 5))
    
    # AUC
    plt.subplot(1, 3, 1)
    plt.plot(epochs, loss, 'r-', label='Training AUC')
    plt.plot(epochs, val_loss, 'b-', label='Validation AUC')
    plt.title('Training and Validation AUC')
    plt.xlabel('Epochs')
    plt.ylabel('AUC')
    plt.legend()
    
    # Loss
    plt.subplot(1, 3, 2)
    plt.plot(epochs, loss, 'r-', label='Training Loss')
    plt.plot(epochs, val_loss, 'b-', label='Validation Loss')
    plt.title('Training and Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    
    # Accuracy
    plt.subplot(1, 3, 3)
    plt.plot(epochs, accuracy, 'r-', label='Training Accuracy')
    plt.plot(epochs, val_accuracy, 'b-', label='Validation Accuracy')
    plt.title('Training and Validation Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    
    plt.show()
```

Second model/ improved model and reasoning

```python
def build_model_2(hp):
    model = keras.Sequential([
        keras.Input(shape=(96, 96, 3)),
        keras.Rescaling(1./255),
        keras.Cropping2D(cropping=32),
        
        keras.Conv2D(32, (3,3), activation='relu'),
        keras.BatchNormalization(),
        keras.MaxPooling2D((2,2)),
        
        keras.Conv2D(64, (3,3), activation='relu'),
        keras.BatchNormalization(),
        keras.MaxPooling2D((2,2)),
        
        keras.Conv2D(128, (3,3), activation='relu'),
        keras.BatchNormalization(),
        keras.MaxPooling2D((2,2)),
        
        keras.Flatten(),
        keras.Dense(hp.Int("dense", 64, 128, step=64), activation='relu'),
        keras.Dropout(0.5),
        keras.Dense(1, activation='sigmoid')
])
    model.compile(optimizer=Adam(hp.Choice("learning_rate", [1e-4, 1e-3])),
                  loss='binary_crossentropy', metrics=['accuracy', 'auc'])
    return model

cnn_2_tuner = kt.RandomSearch(
    build_model_2,
    objective="val_accuracy",
    directory="tuning_2"
)

cnn_2_tuner.search(X, y, validation_split=0.2, epochs=10)
```

```python
cnn_2_tuner.results_summary()
```

```python
cnn_2_model = cnn_2_tuner.get_best_models(num_models=1)[0]
cnn_2_model.summary()
```

```python
X = np.array([load_image(i) for i in train_labels['id']])
y = train_labels['label'].values
# reduce the learning rate when a metric has stopped improving.
learning_rate_reduction = keras.ReduceLROnPlateau(monitor='val_loss', 
                                            factor=0.5, 
                                            patience=5, 
                                            min_lr=1e-5)

earlystop = EarlyStopping(monitor='val_loss', 
                          patience=5, 
                          restore_best_weights=True)
best_hps = cnn_2_tuner.get_best_hyperparameters(num_trials=1)[0]

best_model = build_model_2(best_hps)
h2 = best_model.fit(X, y, 
                    epochs=50, 
                    validation_split=0.2, 
                    callbacks=[learning_rate_reduction, earlystop])
```
